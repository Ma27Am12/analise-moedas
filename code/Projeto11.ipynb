# Limpeza e Transformação


```python
moedas = df.T
display(moedas)
```

```python
moedas.info()
```

```python
moedas.rename (columns= {
    'code':'código',
    'name':'nome',
    'high':'máximo',
    'low':'mínimo',
    'varBid':'variância',
    'pctChange':'percentual',
    'bid':'cotação (compra)',
    'ask':'cotação (venda)',
    'timestamp':'timestamp',
    'create_date':'data de criação'
}, inplace=True)
moedas.head()
```

```python
def to_float(coluna):
  moedas[coluna] = moedas[coluna].astype(float)
```

```python
coluna = 'máximo'
to_float(coluna)
moedas.head()
```

```python
moedas.info()
```

```python
to_float('mínimo')
to_float('cotação (compra)')
to_float('cotação (venda)')
```

```python
to_float('variância')
to_float('percentual')
moedas.info()
```

```python
moedas['data de criação'] = pd.to_datetime(moedas['data de criação'])
moedas.info()
```
